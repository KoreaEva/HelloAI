# URL Document Loader
LangChain은 웹에 기록된 글도 텍스트 형식으로 가져와 LLM에 활용할 수 있습니다. 대표적인 URL Loader는 WebBaseLoader와 UnstructuredURLLoader가 있습니다.

```python
from langchain.document_loaders import WebBaseLoader

loader = WebBaseLoader('https://www.nocutnews.co.kr/news/6062314')
data = loader.load()
print(data[0].page_content)
```

```python
from langchain.document_loaders import Docx2txtLoader
loader = Docx2txtLoader('/content/LLM이 만들어 내는 혁명과 교통산업.docx')

data = loader.load()
```

```python
from langchain.document_loaders.csv_loader import CSVLoader
loader = CSVLoader(file_path='/content/titanic3.csv', csv_args={
    'delimiter': ',',
    'quotechar': '"',
    'fieldnames': ['pclass','survived','name','sex','age','sibsp','parch','ticket','fare','cabin','embarked','boat','body','home.dest']
})

data = loader.load()
```
